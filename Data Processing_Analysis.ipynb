```python
import json
with open('./data/sample_data.json', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame()
for i in range(0, len(data['data'])):
    curr = data["data"][i]
    flat_data = {
        **curr,
        **{f"player_{k}": v for k, v in curr['player'].items()},
        **{f"team_{k}": v for k, v in curr['team'].items()},
        **{f"game_{k}": v for k, v in curr['game'].items()}
    }

    # Remove the original nested dictionaries
    flat_data.pop('player')
    flat_data.pop('team')
    flat_data.pop('game')

    # Create a DataFrame
    df = df.append([flat_data])
df = df.set_index("id")
df
```
