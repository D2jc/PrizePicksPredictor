```python
import numpy as np
import pandas as pd
```

```python
import json
with open('response_2024-02-07.json', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame()
for x in range(0, len(data['data'])):
    curr = data["data"][x]

    for i in range(0, len(curr['home_team']['players'])):
        flat_data = {
            **curr,
            **{f"homeTeam_{k}": v for k, v in curr['home_team'].items() if k != 'players'},
            **curr['home_team']['players'][i],
            **{f"homeTeamPlayer_{k}": v for k, v in curr['home_team']['players'][i]['player'].items()},
        }
        df = pd.concat([df, pd.DataFrame([flat_data])])
    for j in range(0, len(curr['visitor_team']['players'])):
        flat_data_visit = {
            **curr,
            **{f"visitorTeam_{k}": v for k, v in curr['visitor_team'].items() if k != 'players'},
            **curr['visitor_team']['players'][j],
            **{f"visitorTeamPlayer_{k}": v for k, v in curr['visitor_team']['players'][j]['player'].items()},
        }
        df = pd.concat([df, pd.DataFrame([flat_data_visit])])
```
