```python
import json
with open('response_2024-02-07.json', 'r') as file:
    data = json.load(file)
```

```python
homedf = pd.DataFrame()
visitordf = pd.DataFrame()
for x in range(0, len(data['data'])):
    curr = data["data"][x]

    for i in range(0, len(curr['home_team']['players'])):
        flat_data = {
            **curr,
            **{f"Team_{k}": v for k, v in curr['home_team'].items() if k != 'players'},
            **curr['home_team']['players'][i],
            **{f"Player_{k}": v for k, v in curr['home_team']['players'][i]['player'].items()},
        }
        homedf = pd.concat([homedf, pd.DataFrame([flat_data])])
    for j in range(0, len(curr['visitor_team']['players'])):
        flat_data_visit = {
            **curr,
            **{f"Team_{k}": v for k, v in curr['visitor_team'].items() if k != 'players'},
            **curr['visitor_team']['players'][j],
            **{f"Player_{k}": v for k, v in curr['visitor_team']['players'][j]['player'].items()},
        }
        visitordf = pd.concat([visitordf, pd.DataFrame([flat_data_visit])])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming homedf and visitordf are already loaded

# Set the style for better-looking plots
plt.style.use('seaborn')


# 2. Compare team shooting percentages
plt.figure(figsize=(10, 6))
home_percentages = homedf[['fg_pct', 'fg3_pct', 'ft_pct']].mean()
visitor_percentages = visitordf[['fg_pct', 'fg3_pct', 'ft_pct']].mean()

x = ['FG%', '3PT%', 'FT%']
width = 0.35

plt.bar([i - width/2 for i in range(len(x))], home_percentages, width, label='Home Team', color='blue', alpha=0.7)
plt.bar([i + width/2 for i in range(len(x))], visitor_percentages, width, label='Visitor Team', color='red', alpha=0.7)

plt.title('Team Shooting Percentages')
plt.xlabel('Shot Type')
plt.ylabel('Percentage')
plt.ylim(0, 1)
plt.legend()
plt.xticks(range(len(x)), x)
plt.tight_layout()
plt.show()

# 3. Minutes played vs Points scored scatter plot
plt.figure(figsize=(12, 8))

# Convert 'min' to numeric, replacing any non-numeric values with 0
homedf['min_numeric'] = pd.to_numeric(homedf['min'], errors='coerce').fillna(0)
visitordf['min_numeric'] = pd.to_numeric(visitordf['min'], errors='coerce').fillna(0)

# Sort the DataFrames by 'min_numeric'
homedf_sorted = homedf.sort_values('min_numeric')
visitordf_sorted = visitordf.sort_values('min_numeric')

plt.scatter(homedf_sorted['min_numeric'], homedf_sorted['pts'], color='blue', alpha=0.7, label='Home Team')
plt.scatter(visitordf_sorted['min_numeric'], visitordf_sorted['pts'], color='red', alpha=0.7, label='Visitor Team')


plt.title('Minutes Played vs Points Scored')
plt.xlabel('Minutes Played')
plt.ylabel('Points Scored')
plt.legend()
plt.tight_layout()
plt.show()

# 4. Team Rebounds Comparison
plt.figure(figsize=(10, 6))
home_rebounds = homedf[['oreb', 'dreb']].sum()
visitor_rebounds = visitordf[['oreb', 'dreb']].sum()

plt.bar(['Home Team', 'Visitor Team'], [home_rebounds['oreb'], visitor_rebounds['oreb']], label='Offensive Rebounds', color='lightblue')
plt.bar(['Home Team', 'Visitor Team'], [home_rebounds['dreb'], visitor_rebounds['dreb']], bottom=[home_rebounds['oreb'], visitor_rebounds['oreb']], label='Defensive Rebounds', color='darkblue')

plt.title('Team Rebounds Comparison')
plt.ylabel('Number of Rebounds')
plt.legend()
plt.tight_layout()
plt.show()


# Assuming homedf and visitordf are already loaded with data from multiple teams

# Set style for better-looking plots
plt.style.use('seaborn')

# Calculate average statistics for home and visitor teams
home_stats = homedf[['pts', 'reb', 'ast', 'stl', 'blk', 'turnover']].mean()
visitor_stats = visitordf[['pts', 'reb', 'ast', 'stl', 'blk', 'turnover']].mean()

# Prepare data for plotting
stats = ['Points', 'Rebounds', 'Assists', 'Steals', 'Blocks', 'Turnovers']
home_values = home_stats.values
visitor_values = visitor_stats.values

# Set up the plot
fig, ax = plt.subplots(figsize=(14, 8))

# Set the width of each bar and the positions of the bars
width = 0.35
x = np.arange(len(stats))

# Create the bars
rects1 = ax.bar(x - width/2, home_values, width, label='Home Teams', color='blue', alpha=0.7)
rects2 = ax.bar(x + width/2, visitor_values, width, label='Visitor Teams', color='red', alpha=0.7)

# Customize the plot
ax.set_ylabel('Average per Game')
ax.set_title('Home vs Visitor Team Statistics')
ax.set_xticks(x)
ax.set_xticklabels(stats)
ax.legend()

# Add value labels on top of each bar
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()


# Set style for better-looking plots
plt.style.use('seaborn')

# 1. Average Points per Game: Home vs Visitor
plt.figure(figsize=(10, 6))
home_pts = homedf.groupby('Team_name')['pts'].sum().mean()
visitor_pts = visitordf.groupby('Team_name')['pts'].sum().mean()

plt.bar(['Home Teams', 'Visitor Teams'], [home_pts, visitor_pts], color=['blue', 'red'])
plt.title('Average Team Points: Home vs Visitor')
plt.ylabel('Average Points per Game')
for i, v in enumerate([home_pts, visitor_pts]):
    plt.text(i, v, f'{v:.1f}', ha='center', va='bottom')
plt.show()

# 4. Home vs Visitor Win Percentage
home_wins = (homedf['home_team_score'] > homedf['visitor_team_score']).mean()
visitor_wins = (visitordf['visitor_team_score'] > visitordf['home_team_score']).mean()

plt.figure(figsize=(10, 6))
plt.bar(['Home Teams', 'Visitor Teams'], [home_wins, visitor_wins], color=['blue', 'red'])
plt.title('Win Percentage: Home vs Visitor')
plt.ylabel('Win Percentage')
for i, v in enumerate([home_wins, visitor_wins]):
    plt.text(i, v, f'{v:.2%}', ha='center', va='bottom')
plt.ylim(0, 1)
plt.show()

# 5. Box Plot of Points Scored
plt.figure(figsize=(10, 6))
sns.boxplot(data=[homedf['pts']])
plt.title('Distribution of Points Scored: Home')
plt.ylabel('Points')
plt.show()

# 5. Box Plot of Points Scored
plt.figure(figsize=(10, 6))
sns.boxplot(data=[visitordf['pts']])
plt.title('Distribution of Points Scored: Visitor')
plt.ylabel('Points')
plt.show()
```
